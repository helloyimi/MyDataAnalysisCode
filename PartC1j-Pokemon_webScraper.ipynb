```python
import csv
import requests
from bs4 import BeautifulSoup
from IPython.display import HTML
```

```python
# scrape the current Detainees of Boone County Jail from webpage into CSV
url = 'https://www.serebii.net/pokemon/gen1pokemon.shtml'
headers = {'User-Agent': "Chrome/54.0.2840.90"}
response = requests.get(url, headers=headers,verify=False)
html = response.content 
```

```python
HTML('<iframe src=https://www.serebii.net/pokemon/gen1pokemon.shtml width=900 height=350></iframe>')
```

```python
soup = BeautifulSoup(html, "lxml")
table = soup.find('div', attrs={'id': 'content'})
```

```python
tmpRow = (table.findAll('tr')[2:])
print (tmpRow)

# The <tr> tag specifies a row in an HTML table.
# Each table data/cell is defined with a <td> tag.
```

```python
### Editing the code to fetch details of every record. The 'Case Number' and 'Charge Description' is Fetched for every Record.
try:
    outfile = open("./pokemon.csv", "w")  # create an empty file in write mode
    writer = csv.writer(outfile)
    
    # Add header to the file
    writer.writerow(["No.", "Name", "Type", "Abilities", "HP", 'Att', 'Def', 'S.Att', 'S.Def', 'Spd'])
    #load the data
    big = []
    big2 = []
    for row in (table.findAll('tr')[2:]):
        small = []
        #the text data
        for cell in row.findAll("td"):
            text = cell.text.strip()
            small.append(text)
        data = list(filter(None, small))
        if data != []:
            big.append(data)

        #the data from html
        small2=[]
        for anchor in row.findAll("a"):
            type_p = anchor['href'].split("/")
            if "type" in type_p:
                r_type = type_p[-1]
                small2.append(r_type)
        if small2 != []:
            big2.append(small2)
    
    #add the type into the list
    for i in range(len(big2)):
        if len(big2[i]) > 1:
            item = big2[i]
            item = '/'.join(str(n) for n in item)
            big[i].insert(2,item)
        else:
            big[i].insert(2,big2[i][0])
    

        arrLength = len(big[i])
        writer.writerow(big[i])
finally:
    outfile.close() 
```
